```python
import pandas as pd
import git_access
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import networkx as nx
import os
```

```python
project_df = pd.read_csv('1000.csv')
```

```python
project_df['# Issues'] = [0]*project_df.shape[0]
project_df['# Pulls'] = [0]*project_df.shape[0]
project_df['# Commits'] = [0]*project_df.shape[0]
project_df['# contributors'] = [0]*project_df.shape[0]
project_df['# Created_date'] = [0]*project_df.shape[0]
project_df['# last_Updated_date'] = [0]*project_df.shape[0]
```

```python
def get_checks(project_df):
    for i in range(project_df.shape[0]):
        print(i)
        access_token = 'a56313e073db3019ec987bc2d9ea450a1a575450'
        repo_owner = project_df.iloc[i,1]
        source_type = 'github_repo'
        git_url = project_df.iloc[i,0]
        api_base_url = 'http://api.github.com'
        repo_name = project_df.iloc[i,2]
        page_number = 1
        client = git_access.GitHubClient({'access_token': access_token,
                               'repo_owner': repo_owner, 
                               'source_type': source_type, 
                               'git_url': git_url, 
                               'api_base_url': api_base_url, 
                               'repo_name': repo_name}) 
        
        # Issue
        url_type = 'issues'
        print(repo_owner,repo_name)
        base_url = api_base_url + '/repos/' + repo_owner + '/' + str(repo_name) + '/' + url_type
        paged_url = base_url + '?state=' + 'all' + '&page=' + str(page_number) + '&per_page=100'
        res = client.get(paged_url)
        x = json.loads(res.content)
        project_df.iloc[i,3] = len(x)
        
        #Pulls
        url_type = 'pulls'
        base_url = api_base_url + '/repos/' + repo_owner + '/' + str(repo_name) + '/' + url_type
        paged_url = base_url + '?page=' + str(page_number) + '&per_page=100'
        res = client.get(paged_url)
        x = json.loads(res.content)
        project_df.iloc[i,4] = len(x)
        
        #Contributers
        url_type = 'commits'
        base_url = api_base_url + '/repos/' + repo_owner + '/' + str(repo_name) + '/' + url_type
        paged_url = base_url + '?page=' + str(page_number) + '&per_page=100'
        res = client.get(paged_url)
        x = json.loads(res.content)
        project_df.iloc[i,5] = len(x)
        
        #Contributers
        url_type = 'contributors'
        base_url = api_base_url + '/repos/' + repo_owner + '/' + str(repo_name) + '/' + url_type
        paged_url = base_url + '?page=' + str(page_number) + '&per_page=100'
        res = client.get(paged_url)
        x = json.loads(res.content)
        project_df.iloc[i,6] = len(x)
        
        #Duration
        base_url = api_base_url + '/repos/' + repo_owner + '/' + str(repo_name)
        paged_url = base_url
        res = client.get(paged_url)
        x = json.loads(res.content)
        project_df.iloc[i,7] = x['created_at']
        project_df.iloc[i,8] = x['updated_at']
```

```python
get_checks(project_df)
```
