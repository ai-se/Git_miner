```python
import pandas as pd
```

```python
df = pd.read_pickle('hadoop_committed_file.pkl')

commits = df.commit_id.unique()
```

```python
for commit in commits:
    for index,items in df[df['commit_id'] == commit].iterrows():
        if len(items['file_path'].split('src/')) == 1:
            continue
        x = items['file_path'].split('src/')[1].replace('/','.').rsplit('.',1)
        if x[1] == 'java':
            print(x[0])
```
